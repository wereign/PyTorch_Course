```python
from fileinput import filename


path = './CATS_DOGS/'

img_names = []

for folder,subfolders,filenames in os.walk(path):
    for img in filenames:
        img_names.append(folder+'/'+img)  # Getting all the images from the folders


print(len(img_names))
```

```python
img_sizes = []
rejected = []


for item in img_names:
    try:
        with Image.open(item) as img:
            img_sizes.append(img.size)  # Getting info on the size of each image
    
    except:
        rejected.append(item)
```

```python
print(len(img_sizes))
print(len(rejected))
```

```python
df = pd.DataFrame(img_sizes)
df.head()
```

# Image Preprocessing and various transformations

```python
dog = Image.open('./CATS_DOGS/train/DOG/14.jpg')
display(dog)
```

```python
dog.size
```

```python
dog.getpixel((250,163))   # Gets the RGB value of each pixel
```

```python
transform = transforms.Compose([
    transforms.ToTensor()
])    # Can provide a list of transformations to do seuqentially

im = transform(dog)
print(type(im))
print(im.shape)
```

```python
# matplot lib wants 386,500,3
# (1,2,0)
plt.imshow(np.transpose(im.numpy(),(1,2,0)))
```

```python
print(im)
```

```python
im[:,0,0]   # Values for RGb are normalized, between 0 and 1
```

```python
# Resizing an image
transform = transforms.Compose([
    transforms.Resize(250), # Will make the smaller edge match the image. Passing in tuple matches both axes
    transforms.CenterCrop(250),
    transforms.ToTensor()
])    # Can provide a list of transformations to do seuqentially

im = transform(dog)
print(type(im))
print(im.shape)
plt.imshow(np.transpose(im.numpy(),(1,2,0)))
```

```python
# Resizing an image
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(0.5),
    transforms.ToTensor()
])    # Can provide a list of transformations to do seuqentially

im = transform(dog)
print(type(im))
print(im.shape)
plt.imshow(np.transpose(im.numpy(),(1,2,0)))
```

```python
# Resizing an image
transform = transforms.Compose([
    transforms.RandomRotation(30),
    transforms.ToTensor()
])    # Can provide a list of transformations to do seuqentially

im = transform(dog)
print(type(im))
print(im.shape)
plt.imshow(np.transpose(im.numpy(),(1,2,0)))
```

```python
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=1),
    transforms.RandomRotation(30),
    transforms.Resize(224),
    transforms.CenterCrop(224),
    transforms.ToTensor()])

im = transform(dog)
print(type(im))
print(im.shape)
plt.imshow(np.transpose(im.numpy(),(1,2,0)))
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406],std=[0.229,0.224,0.225])
])


im = transform(dog)
print(type(im))
print(im.shape)
plt.imshow(np.transpose(im.numpy(),(1,2,0)))
```

```python
inv_normalize = transforms.Normalize(
mean=[-0.485/0.229,-0.456/0.224,-0.406/0.225],std=[1/0.229,1/0.224,1/0.225])

im_inv = inv_normalize(im)
plt.figure(figsize=(12,4))
plt.imshow(np.transpose(im_inv.numpy(),(1,2,0)))
```
