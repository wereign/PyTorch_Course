````python
import torch
import torch.nn as nn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
# Preparing the training data 
x = torch.linspace(0,799,800)
print(x)
````

````python
y = torch.sin(2*x*3.146/40)
````

````python
test_size = 40
train_set = y[:-test_size]
test_set = y[-test_size:]
````

````python
plt.figure(figsize=(12,4))
plt.xlim(-10,801)
plt.grid(True)
plt.plot(test_set.numpy())
````

````python
# Prepare the training sequence

def input_data(seq,ws):
    out = []  # ([1,2,3],[4]) --> ([1,2,3,4],[5])
    L = len(seq) # TO ensure the internal split doesn't go beyond the last point
    
    for i in range(L-1): # TO not reach the last element
        window = seq[i:i+ws]
        label = seq[i+ws:i+ws+1]
        out.append((window,label))
    return out
````

````python
window_size = 40
train_data = input_data(train_set,window_size)
````

Setting up the LSTM Model
1. Long Term and Short Term Memory
2. Input at time t
3. New Long Term and Short Term Memory
4. Output at time t
```Python
torch.nn.LSTMCell(input_size,hidden_size,bias=True)
```

- bias - true by default
- Input - (input,(long term memory, short term memory))
- Similar Pattern for output

````python
class LSTM(nn.Module):
    def __init__(self,input_size=1,hidden_size=50,out_size=1):
        super().__init__()
        
        self.hidden_size = hidden_size
        
        
        # Setting up LSTM
        self.lstm = nn.LSTM(input_size,hidden_size)
        self.linear = nn.Linear(hidden_size,out_size)
        
        
        self.hidden = (torch.zeros(1,1,hidden_size),torch.zeros(1,1,hidden_size))
        
        
    
    def forward(self,seq):
        
        
        lstm_out, self.hidden = self.lstm(seq.view(len(seq),1,-1),self.hidden)
        
        
        pred = self.linear(lstm_out.view(len(seq),-1))
        
        #  prediction will give the entire sequence including 
        
        return pred[-1]
    
        
        
````

````python
torch.manual_seed(42)
model = LSTM()
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
````

- Test range - 760 to 800
- Once we are satisfied with the results on the test portion, it will be time to forecast into the unknown feature
- Then retrain on all available data
- In real life, there would be no values to compare the predictions too

````python
train_set[-window_size:].tolist()
````

````python
epochs = 10
future  = 40

for i in range(epochs):
    
    for seq,y_train in train_data:
        optimizer.zero_grad()
        model.hidden = (torch.zeros(1,1,model.hidden_size),torch.zeros(1,1,model.hidden_size))
        
        y_pred = model(seq)
        loss = criterion(y_pred,y_train)
        loss.backward()
        optimizer.step()
    
    print(f"Epoch {i} Loss {loss.item()}")
    
    # Plotting predictions against test set range
    
    preds = train_set[-window_size:].tolist()
    
    for f in range(future):
        seq = torch.FloatTensor(preds[-window_size:])
        
        with torch.no_grad():
            model.hidden = (torch.zeros(1,1,model.hidden_size),torch.zeros(1,1,model.hidden_size))
            
            preds.append(model(seq).item())
            
    loss = criterion(torch.tensor(preds[-window_size:]),y[760:])
    print(f'Test Range Loss: {loss}')
    
    
    plt.figure(figsize=(12,4))
    plt.xlim(700,801)
    plt.grid(True)
    plt.plot(y.numpy())
    plt.plot(range(760,800),preds[-window_size:])
    plt.show()
````
